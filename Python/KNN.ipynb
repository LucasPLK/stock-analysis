```python
import pandas as pd
```

```python
from sklearn.neighbors import KDTree
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('../data/data_cleaned_fillna.csv', index_col=0)
```

```python
data.columns
```

```python
scaler = StandardScaler()
```

```python
columns = data.drop(columns=['Ticker', 'Setor'], axis=0).columns
data_scaled = data.loc[:, ['Ticker', 'Setor']]
                        
data_scaled[columns] = scaler.fit_transform(data.drop(columns=['Ticker', 'Setor'], axis=0))
```

```python
data_scaled
```

```python
tree = KDTree(data_scaled.drop(columns=['Ticker', 'Setor'], axis=0))
```

```python
dist, ind = tree.query(data_scaled.loc[data_scaled['Ticker'] == 'WEGE3'].drop(columns=['Ticker', 'Setor'], axis=0),\
                       k=6)  
```
